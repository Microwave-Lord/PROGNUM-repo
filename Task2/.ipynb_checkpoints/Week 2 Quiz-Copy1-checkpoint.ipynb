```python
#Question 2.12

from math import atan2, degrees
x = float(input("value for x?: "))
y = float(input("value for y?: "))

n = atan2(y, x)

if 0 <= n <= 3.14/2:
    print("1st")
elif 3.14/2 < n <= 3.14:
    print("2nd")
elif -3.14/2 <= n <0:
    print("4th!")
else:
    print("3rd!")
```
