```python
def number():
    a = float(input('number'))
    return a

def op():
    b = str(input('operation'))
    return b

def main2():
    z = op()
    if z == '+':
        b = number()
        c = a + b
        a = c
        print(c)
    elif z == '-':
        b = number()
        c = a - b
        a = c
        print(c)
    

def main():
    x = number()
    z = op()
    y = number()
    if z == '+':
        a = x+y
        print(a)
        main2()
    if z == '-':
        a = x-y
        print(a)
        main2()
main()
```
