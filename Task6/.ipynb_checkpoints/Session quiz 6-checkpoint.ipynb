```python
#question 6.1
from astropy.io import fits
import numpy as np
from scipy import stats as st
import matplotlib.pyplot as plt
hdu = fits.open('m101 (1).fits')

#there are 2 HDUs
#PRIMARY has the image data, index = [0], has number type int16 which means it can handle numbers from -32768 to 32768!
print(hdu.info())

header = hdu[0].header
for i in header[:5]:
    print(f'{i:18s} = {header[i]}')
#data type does correspond!


dat = hdu[0].data
print(f'{np.shape(dat)} is the shape!') #this shape does  match what is in info, however it does not correspond to the NAXIS header, only the NAXIS1 header!
print(dat.dtype) #the dtype is >i2, which does match the int16 type of before; >i2 means a 'big-endian' data with 2-byte signed values
#1 byte = 2bits, so 2bytes = 16bit

dt = dat.flatten()
dt_min = np.min(dt)
dt_max = np.max(dt)
dt_mean = np.mean(dt)
dt_std = np.std(dt)
dt_med = np.median(dt)        
mode = st.mode(dt)[0]
print(f'{dt_min} is the min, {dt_max} is the max, {dt_mean} is the mean, {dt_std} is the std dev, {dt_med} is the median, {mode} is the mode!')

fig, ax = plt.subplots(1)
ax.hist(dt, bins = 30)
ax.axvline(dt_mean, color = 'green', label = 'mean')
ax.axvline(dt_med, color = 'yellow', label = 'median')
ax.axvline(mode, color = 'red', label = 'mode')
ax.legend()


fig2, ax2 = plt.subplots(1)
data = np.asarray(dat)
median2 = np.median(data)
CTYPE1 = header['CTYPE1']
CTYPE2 = header['CTYPE2']

bruh = ax2.imshow(data, cmap = 'Blues', vmin = median2)
ax2.set_xlabel(CTYPE1)
ax2.set_ylabel(CTYPE2)
fig2.colorbar(bruh, cmap = 'Blues', label = 'Counts')


header['BUNIT'] = 'Counts'

hdulistout = fits.PrimaryHDU(data)
hdulistout.header = header   # We can copy the old header
hdulistout.writeto('newdata.fits', overwrite=True, output_verify='ignore')
```

```python
from astropy.io import fits
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
import math as m

hdu = fits.open('m101 (1).fits')
fig, ax = plt.subplots()
data = hdu[0].data



c = np.percentile(data, 70)
ax.imshow(data, vmax = c)

d_s = np.array(data[20:40, 20:40])
d = d_s.astype(float)
rms = m.sqrt((1/400)*(np.sum(d**2)))
print(rms, np.std(d_s))
#Rms is way higher yolo 
```

```python
from astropy.io import fits
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
import math as m

n1 = fits.open('noisemap0.fits')
n2 = fits.open('noisemap1.fits')
n3 = fits.open('noisemap2.fits')
data_0 = n1[0].data
data_1 = n2[0].data
data_2 = n3[0].data


std1, std2, std3 = np.std(data_0), np.std(data_1), np.std(data_2)
print(std1, std2, std3)
stack = np.stack((data_0, data_1, data_2), axis = 0)

baz = np.mean(stack, axis = 0)
baz_std = np.std(baz)
std = (1/3)*m.sqrt(std1**2 + std2**2 + std3**2)

print(f'{baz_std} and {std} match')
fig, ax = plt.subplots()

ax.imshow(baz)
```

```python
#Question 6.6
from astropy.io import fits
import numpy as np
from astropy.table import Column, Table

# Open the FITS file
hdul = fits.open('APOGEE_stars.fits')


# there is no difference in reading th data using astropy.io

# Access the primary data (in this case, it is a Table)

data = hdul[1].data  # Often, the first extension contains the table
colnames = hdul[1].columns.names # List of the column names
print(hdul.info()) #there is a difference in the type, the dimensions and the format compared to the first header which would be the image data


mask = (data['STARFLAG'] == 0) & (data['ASPCAPFLAG'] == 0) & (data['SNR'] > 10) & ((data['GAIAEDR3_PARALLAX']/data['GAIAEDR3_PARALLAX_ERROR']) >5) & (~np.isnan(data['GAIAEDR3_PHOT_G_MEAN_MAG'])) & (~np.isnan(data['GAIAEDR3_PHOT_BP_MEAN_MAG'])) & (~np.isnan(data['GAIAEDR3_PHOT_RP_MEAN_MAG']))


#makes the DATAFLAG column
c = []
for i in range (len(data)):
    if mask[i] == False:
        c.append(0)
    else:
        c.append(1)


#this was to check that our data was being processed
print(np.shape(data), len(c))

data_t = Table(data, names =(colnames))
data_t.add_column(c, name = 'DATAFLAG', index = -1)

MG = (data_t['GAIAEDR3_PHOT_G_MEAN_MAG']+5*np.log10(data_t['GAIAEDR3_PARALLAX']/1000)+5)

#looping through the column to replace GAIAEDR3 to replace relative with absolute


mask2 = (data_t['DATAFLAG'] == 0)
Abs_Mag = []


for j in range (len(data_t['GAIAEDR3_PHOT_G_MEAN_MAG'])):
    if mask2[j] == False:
        Abs_Mag.append(-1)
    else:
        Abs_Mag.append(MG[j])
        
data_t.add_column(Abs_Mag, name = "ABSOLUTE MAGNITUDE", index = -1)

print(data_t)

data_t.write('APOGEE_update.fits', overwrite=True)
```

```python
import math as m
import numpy as np

class Fibonacci:
    def __init__(self, n, m):
        self.n = n
        self.m = m
    def nth(self):
        a = 0
        b = 1
        for i in range (2, (self.n)):
            c = a + b
            a = b
            b = c
        return b
    def fibo(self):
        fiblist = []
        a = 0
        b = 1
        l = 2
        while l<(self.n)-1:
            c = a+b
            a = b
            b = c
            if a%(self.m) == 0:
                fiblist.append(a)
            l = l+1
        return fiblist
    
    
fib = Fibonacci(100, 7)
print(fib.nth())
print(fib.fibo())
```
