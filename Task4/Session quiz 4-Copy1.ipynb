```python
#Question 4.4
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

#generates distribution
length = np.random.normal(loc = 185, scale = 8, size = 100000)
print(len(length))
a = np.std(length)
b = np.mean(length)


#calculates probability that a given height is below z
z = float(input("what height?"))
cdf = norm.cdf(z, scale = 8, loc = 185)
print(f"{cdf*100}% is the probability that the height is below {z}")


#calculates the percentage of data in one standard deviation to the left and right
Norm1 = norm.cdf(a+b, scale = 8, loc = 185)
Norm2 = norm.cdf(b-a, scale = 8, loc = 185)
print(f"{100*(Norm1 - Norm2)} is the percentage of data within one standard deviation")


#plots everything
plt.hist(length, bins = 40, label = str(f"std:{a}, mean:{b}"))
plt.axvspan(b-a, b+a, color = 'red', alpha = 0.2)
plt.vlines(b, 0, 9000, color = 'red')
plt.legend()
plt.show()

```

```python
#4.21

import numpy as np 
from matplotlib import pyplot as plt
from scipy.stats import norm, multivariate_normal



x = (np.random.normal(loc = 0, scale = 1, size = 100000))
y = (np.random.normal(loc = 2, scale = 1, size = 100000))


x_1 = np.linspace(-6, 8, 30)
x_new = norm.pdf(x_1, scale=1, loc=0)

y_1 = np.linspace(-2, 6.5, 30)
y_new = norm.pdf(y_1, scale = 1, loc = 2)



plt.plot(x_1, x_new, color = 'black', label = 'X(PDF)', linestyle = ':')
plt.plot(y_1, y_new, color = 'black', label = 'Y(PDF)', linestyle = '--')
plt.hist(x, bins = 30, color = 'blue', alpha = 0.4, density = True, label = 'X')
plt.hist(y, bins = 30, color = 'red', alpha = 0.4, density = True, label = 'Y')
plt.grid(axis = 'y')
plt.legend()
plt.xlabel('Values')
plt.xlim(-6, 8)
plt.ylabel('Probability Density')



# Blue arrow pointing right, label placed left outside
plt.annotate(
    r'$\mathcal{N}(0, 1)$',
    xy=(-2, 0.15),           # arrow tip: close to curve on left
    xytext=(-5, 0.15),       # text far left
    color='blue',
    fontsize=14,
    arrowprops=dict(facecolor='blue', shrink=0.05, width=3, headwidth=12, headlength=10)
)

# Red arrow pointing left, label placed right outside
plt.annotate(
    r'$\mathcal{N}(2, 1)$',
    xy=(4.8, 0.15),            # arrow tip: close to curve on right
    xytext=(7, 0.15),          # text far right
    color='red',
    fontsize=14,
    arrowprops=dict(facecolor='red', shrink=0.05, width=3, headwidth=12, headlength=10)
)
plt.show()




# 2D histogram
counts, xedges, yedges, im = plt.hist2d(x, y, bins=50, cmap='Blues')

# Convert to PDF
pdf = counts / np.sum(counts)

# Flatten and sort in descending order
pdf_flat = pdf.flatten()
idx = np.argsort(pdf_flat)[::-1]
pdf_sorted = pdf_flat[idx]

# CDF
cdf = np.cumsum(pdf_sorted)

# Find density levels for 1σ, 2σ, 3σ
levels = []
for p in [0.6827, 0.9545, 0.9973]:
    level = pdf_sorted[np.searchsorted(cdf, p)]
    levels.append(level)

levels = np.sort(levels)
counts_percent = 100 * counts / counts.max()

plt.imshow(
    counts_percent.T,
    origin='lower',
    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
    cmap='Blues',
    vmin=0,
    vmax=100
)

cbar = plt.colorbar()




# Overlay contours (use bin centers!)
xcenters = 0.5 * (xedges[:-1] + xedges[1:])
ycenters = 0.5 * (yedges[:-1] + yedges[1:])

CS = plt.contour(
    xcenters,
    ycenters,
    pdf.T,
    levels=levels,
    colors='red'
)

fmt = {
    levels[0]: "3σ",
    levels[1]: "2σ",
    levels[2]: "1σ"
}

plt.clabel(CS, CS.levels, fmt=fmt, inline= True, fontsize = 12)
plt.xlabel('x')
plt.ylabel('y')
plt.title('2d histogram + contour')
plt.grid(linestyle = '--')
plt.show()
```
