```python
import  numpy as np
from matplotlib import pyplot as plt
import scipy as scp
import math as m 

x = np.linspace(-10, 10, 200)

A = float(input("whats A?"))
x0 =float(input("whats x0?"))
sig = float(input("whats sigma?"))
z0 = float(input("whats z0?"))
a = float(input("whats lower bound?"))
b = float(input("whats upper bound?"))

def gauss(x, A, x0, sigma, z0):
    return A*np.exp(-(x-x0)**2/(2*sigma**2))+z0

aa, error = scp.integrate.quad(lambda x:(A*np.exp(-(x-x0)**2/(2*sig**2))+z0), a, b)


print(aa)

plt.scatter(x, gauss(x,A,x0, sig, z0), label = f'area = {aa}')
plt.fill_between(x, gauss(x,A,x0, sig, z0), where = (x >= a) & (x <=b))
plt.legend()
plt.show()
```
