```python
#4.21

import numpy as np 
from matplotlib import pyplot as plt
from scipy.stats import norm, multivariate_normal



x = (np.random.normal(loc = 0, scale = 1, size = 100000))
y = (np.random.normal(loc = 2, scale = 1, size = 100000))


x_1 = np.linspace(-3, 3, 30)
x_new = norm.pdf(x_1, scale=1, loc=0)

y_1 = np.linspace(-1, 5, 30)
y_new = norm.pdf(y_1, scale = 1, loc = 2)



plt.plot(x_1, x_new, color = 'black', label = 'X(PDF)', linestyle = ':')
plt.plot(y_1, y_new, color = 'black', label = 'Y(PDF)', linestyle = '--')
plt.hist(x, bins = 30, color = 'blue', alpha = 0.4, density = True, label = 'X')
plt.hist(y, bins = 30, color = 'red', alpha = 0.4, density = True, label = 'Y')
plt.grid(axis = 'y')
plt.legend()
plt.xlabel('Values')
plt.xlim(-6, 8)
plt.ylabel('Probability Density')
plt.show()


# 2D histogram
counts, xedges, yedges, im = plt.hist2d(x, y, bins=30, cmap='Blues')

# Convert to PDF
pdf = counts / np.sum(counts)

# Flatten and sort in descending order
pdf_flat = pdf.flatten()
idx = np.argsort(pdf_flat)[::-1]
pdf_sorted = pdf_flat[idx]

# CDF
cdf = np.cumsum(pdf_sorted)

# Find density levels for 1σ, 2σ, 3σ
levels = []
for p in [0.6827, 0.9545, 0.9973]:
    level = pdf_sorted[np.searchsorted(cdf, p)]
    levels.append(level)

levels = np.sort(levels)
counts_percent = 100 * counts / counts.max()

plt.imshow(
    counts_percent.T,
    origin='lower',
    extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]],
    cmap='Blues',
    vmin=0,
    vmax=100
)

cbar = plt.colorbar()




# Overlay contours (use bin centers!)
xcenters = 0.5 * (xedges[:-1] + xedges[1:])
ycenters = 0.5 * (yedges[:-1] + yedges[1:])

plt.contour(
    xcenters,
    ycenters,
    pdf.T,
    levels=levels,
    colors='red'
)

plt.show()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Plot contours from an 2D histogram showing the standard deviation
"""

import numpy as np
import matplotlib.pyplot as plt

# Build datas ###############

x, y = np.random.normal(size=(2, 1000000))

xbins = np.linspace(-2, 2, 30)
ybins = np.linspace(-2, 2, 30)

counts, xedges, yedges = np.histogram2d(x, y, bins=(xbins, ybins))

print("std(x)=", np.std(x))
print("std(y)=", np.std(y))

# Plot data #################

fig, ax = plt.subplots()

sigmas = [1., 2., 3.]
levels = []
fmt = {}

for sigma in sigmas:
    levels.append(float(sigma) * np.std(counts))
    fmt[float(sigma) * np.std(counts)] = r"${}\sigma$".format(int(sigma))

cs = plt.contour(xedges[:-1], yedges[:-1], counts.T, levels,
                  linewidths=(2, 2, 3), linestyles=('dotted', 'dashed', 'solid'),
                  alpha=0.8, colors='red')
ax.clabel(cs, inline=True, fontsize=16, fmt=fmt)

# Set title and labels ######

ax.set_title("Contour", fontsize=20)
ax.set_xlabel(r"$X_1$", fontsize=20)
ax.set_ylabel(r"$X_2$", fontsize=20)

# Set legend ################

lines = [ cs.collections[0]]
labels = [r'$\mathcal{N}$']
ax.legend(lines, labels, prop={'size': 14}, loc='best', fancybox=True, framealpha=0.5)

# Save file #################

plt.savefig("contour_from_hist2d_sigmas.png")

# Plot ######################

plt.show()
```
