```python
#Question 3.2

import math as m
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

c = []
for i in range (len(names)):
    print(f"{masses[i]:.3} is the mass of {names[i]}")
    d = m.sqrt((2*G*masses[i])/(radii[i]))
    c.append(d)

print(c)

esc_v = dict(zip(names, c))
print(esc_v)
```

```python
#Question 3.7

import math as m


def size(d,D):
    c = (2*m.atan((d)/(2*D)))*(180/m.pi)
    print(c)
    
d = float(input("size of the object?"))
D = float(input("Distance of the object?"))

size(d,D)

```

```python
#Questiopn 3.10

import math as m

def av(r):
    A = 2*m.pi*r**2 + 2*m.pi*r*h
    V = m.pi*r**2*h
    return A,V


r = float(input("radius?"))
h = float(input('height?'))

A, V = av(r)
print(f"{A} is the area")
print(f"{V} is the volume")
print(f"{A/V} is the ratio of the area to the volume")
```
