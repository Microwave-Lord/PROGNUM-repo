```python
#Question 3.2

import math as m
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

c = []
for i in range (len(names)):
    print(f"{masses[i]:.3f} is the mass of {names[i]}")
    d = m.sqrt((2*G*masses[i])/(radii[i]))
    c.append(d)



esc_v = dict(zip(names, c))
print(esc_v)
```

```python
import matplotlib.pyplot as plt
import math as m
import numpy as np

x = []
y = []
i = -4

while -6 < i < 6:
    x.append(i)
    c = m.cosh(i)
    y.append(c)
    i = i+1
    

print(y, x)

plt.plot(x, y, marker='+', label = "Catenary")
plt.title("Cosh(x) vs x")
plt.ylabel('cosh(x)')
plt.xlabel('x')
plt.legend()
plt.grid()
plt.show()


a = np.arange(-4, 6)

plt.plot(a, y, marker='+', label = "Catenary")
plt.title("Cosh(a) vs a", fontsize = 16)
plt.legend()
plt.ylabel('cosh(a)')
plt.xlabel('a')
plt.legend
plt.grid()
plt.show()
```
